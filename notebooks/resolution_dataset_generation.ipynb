```python
from PIL import Image

def cambiar_resolucion(img_path, nueva_res):
    img = Image.open(img_path)
    img_resized = img.resize(nueva_res, Image.BILINEAR)
    return img_resized
```

```python
import os

def generar_base_resolucion(dataset_in, dataset_out, nueva_res):
    os.makedirs(dataset_out, exist_ok=True)

    for clase in os.listdir(dataset_in):
        ruta_clase_in = os.path.join(dataset_in, clase)
        ruta_clase_out = os.path.join(dataset_out, clase)

        if not os.path.isdir(ruta_clase_in):
            continue

        os.makedirs(ruta_clase_out, exist_ok=True)

        for nombre_img in os.listdir(ruta_clase_in):
            ruta_img = os.path.join(ruta_clase_in, nombre_img)

            try:
                img_resized = cambiar_resolucion(ruta_img, nueva_res)
                img_resized.save(os.path.join(ruta_clase_out, nombre_img))
            except:
                continue

        print(f"Clase procesada: {clase}")
```

```python
dataset_original = "/content/drive/MyDrive/UCMerced_LandUse/Images"
dataset_32 = "/content/drive/MyDrive/UCMerced_32"

generar_base_resolucion(dataset_original, dataset_32, (32, 32))
```
