```python
import os
import cv2
import numpy as np
```

```python
def awgn_snr(img, snr_db):
    """
    Aplica ruido blanco gaussiano (AWGN) controlado por SNR.
    Retorna imagen uint8 en rango [0,255].
    """

    img_float = img.astype(np.float32)

    # Potencia de la señal
    potencia_signal = np.mean(img_float ** 2)

    # Potencia del ruido según SNR
    potencia_ruido = potencia_signal / (10 ** (snr_db / 10))

    sigma = np.sqrt(potencia_ruido)

    # Generar ruido
    ruido = np.random.normal(0, sigma, img_float.shape).astype(np.float32)

    img_ruido = img_float + ruido

    # Evitar desbordamientos
    img_ruido = np.clip(img_ruido, 0, 255)

    return img_ruido.astype(np.uint8)

```

```python
# Base de datos original
dataset_in = "/content/drive/MyDrive/UCMerced_LandUse/Images"

# Base de datos con ruido
dataset_out = "/content/drive/MyDrive/base_de_datos_con_ruido5dB"
```

```python
snr_db = 5  # SNR en dB
os.makedirs(dataset_out, exist_ok=True)
```

```python
for clase in os.listdir(dataset_in):

    ruta_clase_in = os.path.join(dataset_in, clase)
    ruta_clase_out = os.path.join(dataset_out, clase)

    if not os.path.isdir(ruta_clase_in):
        continue

    os.makedirs(ruta_clase_out, exist_ok=True)

    for nombre_img in os.listdir(ruta_clase_in):

        ruta_img = os.path.join(ruta_clase_in, nombre_img)

        img = cv2.imread(ruta_img)

        if img is None:
            print(f"No se pudo leer: {ruta_img}")
            continue

        # Aplicar AWGN
        img_ruido = awgn_snr(img, snr_db)

        # Guardar como PNG para evitar errores TIFF
        nombre_sin_ext = os.path.splitext(nombre_img)[0]
        ruta_guardado = os.path.join(ruta_clase_out, nombre_sin_ext + ".png")

        cv2.imwrite(ruta_guardado, img_ruido)

    print(f"Clase procesada: {clase}")

print("Base de datos convertida correctamente.")
```
